```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("California_Houses.csv")
```

```python
data.head(10)
```

```python
data.info()
```

```python
label=data["Median_House_Value"]
features=data.drop("Median_House_Value",axis=1)

```

```python
label.head()
```

```python
features.head()
```

```python
from sklearn.model_selection import train_test_split

Features_train,Features_temp,Lables_train,Lables_temp=train_test_split(features,label,test_size=0.3)

Features_valid,Features_test,Lables_valid,Lables_test=train_test_split(features,label,test_size=0.5)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
linear_reg=LinearRegression()
linear_reg.fit(Features_train,Lables_train)
```

```python
lasso_reg=Lasso()
lasso_reg.fit(Features_train,Lables_train)
```

```python
Ridge_reg=Ridge()
Ridge_reg.fit(Features_train,Lables_train)
```

```python
pred_linear=linear_reg.predict(Features_valid)
print(pred_linear)
```

```python
pred_lasso=lasso_reg.predict(Features_valid)
print(pred_lasso)
```

```python
pred_ridge=Ridge_reg.predict(Features_valid)
print(pred_ridge)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
meansqured_linear=mean_squared_error(Lables_valid,pred_linear)
meanabsolute_linear=mean_absolute_error(Lables_valid,pred_linear)

meansqured_lasso=mean_squared_error(Lables_valid,pred_lasso)
meanabsolute_lasso=mean_absolute_error(Lables_valid,pred_lasso)

meansqured_ridge=mean_squared_error(Lables_valid,pred_ridge)
meanabsolute_ridge=mean_absolute_error(Lables_valid,pred_ridge)


print("mean squred error for linear:",meansqured_linear)
print("mean absolute error for linear:",meanabsolute_linear)

print("\nmean squred error for lasso:",meansqured_lasso)
print("mean absolute error for lasso:",meanabsolute_lasso)

print("\nmean squred error for ridge:",meansqured_ridge)
print("mean absolute error for ridge:",meanabsolute_ridge)
```
