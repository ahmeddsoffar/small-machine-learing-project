```python
import numpy as np ## da le el linear algebra
import pandas as pd ##da le el data processing we anee a5od el data mn el file
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
data=pd.read_csv("magic04.data",header=None)
data.head
```

```python
gamma_data = data[data[10] == 'g']  # Assuming 'Class' is the column for labels
hadron_data = data[data[10] == 'h']

# Randomly sample to balance the dataset
gamma_data_balanced = gamma_data.sample(n=len(hadron_data), random_state=42)

# Combine the balanced gamma data with hadron data
balanced_data = pd.concat([gamma_data_balanced, hadron_data])

# Step 2: Splitting the dataset into features and labels
Features = balanced_data.drop([10], axis=1)  # Features
Lables = balanced_data[10]  # Labels
```

```python
Features_train,Features_temp,Lables_train,Lables_temp=train_test_split(Features,Lables,test_size=0.3)

Features_valid,Features_test,Lables_valid,Lables_test=train_test_split(Features_temp,Lables_temp,test_size=0.5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 6)
knn_model.fit(Features_train,Lables_train)
```

```python
predict_valid=knn_model.predict(Features_valid)
print("predication for k=5 :",predict_valid)

accuracy=accuracy_score(predict_valid,Lables_valid)
print("accracy for k=5 :",accuracy)

precision= precision_score(Lables_valid,predict_valid,pos_label="g")
print("precision for k=5:",precision)

f1 = f1_score(Lables_valid,predict_valid,pos_label="g")
print("f1 for k=5:",f1)

recall=recall_score(Lables_valid,predict_valid,pos_label="g")
print("recall for k=5",recall)

cm=confusion_matrix(Lables_valid,predict_valid,labels=["g","h"])
cm_df = pd.DataFrame(cm, index=["Actual Gamma (g)", "Actual Hadron (h)"], columns=["Predicted Gamma (g)", "Predicted Hadron (h)"])
print("\n",cm_df)
```
